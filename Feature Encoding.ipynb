# One Hot Encoding

# Ordinal Encoding

```python
import category_encoders as ce
import pandas as pd
```

```python
df=pd.DataFrame({'Degree':['School','Masters','Diploma','Bachelors','Bachelors','Masters','Phd','School','School']})

#Original data
df
```

```python
df['Degree'].unique()                                                  #Identifies the unique values and returns it.
```

```python
# create object of Ordinalencoding
encoder= ce.OrdinalEncoder(cols=['Degree'],return_df=True,
                 mapping=[{'col':'Degree','mapping':{'None':0,'School':1,'Diploma':2,'Bachelors':3,'Masters':4,'Phd':5}}])
```

```python
df = encoder.fit_transform(df)
```

```python
df
```
